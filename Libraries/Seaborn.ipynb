```python
import seaborn as sb
```

```python
data=sb.load_dataset('tips')
data.head()
```

```python
data.describe()
```

```python
data.corr()
```

```python
# always accepts numerical data
sb.heatmap(data.corr())
```

```python
sb.jointplot(x='tip',y='total_bill',data=data,color='green',kind='hex')
```

```python
sb.pairplot(data,hue='sex',palette='Greens')
```

```python
sb.distplot(data['tip'],bins=20,color='green')
```

```python
sb.countplot('smoker',data=data,color='green')
```

```python
sb.barplot(x='tip',y='day',data=data)
```

```python
sb.boxplot(x='tip',y='day',data=data,hue='sex',palette='Greens')
```

```python
sb.violinplot(x="smoker",y="total_bill",data=data,hue='day',color='green')
```
