```python
# Univariate Analysis
setosa=df.loc[df['species']=='setosa']
virginia=df.loc[df['species']=='virginia']
versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(setosa['sepal_length'],np.zeros_like(setosa['sepal_length']),'*')
plt.plot(virginia['sepal_length'],np.zeros_like(virginia['sepal_length']),'*')
plt.plot(versicolor['sepal_length'],np.zeros_like(versicolor['sepal_length']),'*')
plt.xlabel('Sepal length of different species')
```

```python
# Bivariate analysis
sns.FacetGrid(df,hue='species',height=10).map(plt.scatter,'sepal_length','sepal_width').add_legend();
```

```python
# Multivariate analysis
sns.pairplot(df,hue='species')
```
