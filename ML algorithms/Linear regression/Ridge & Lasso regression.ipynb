```python
from sklearn.datasets import load_boston
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
df=load_boston()
df
```

```python
dataset=pd.DataFrame(df.data)
dataset.head()
```

```python
dataset.columns=df.feature_names
dataset.head()
```

```python
df.target.shape
```

```python
# creating a column for dependent variable
dataset['Prices']=df.target
dataset.head()
```

```python
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,:-1]
```

```python
# Linear regression
lin=LinearRegression()
mse=cross_val_score(lin,X,y,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
mean_mse
```

```python
# Ridge regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regressor=GridSearchCV(ridge,params,X,y,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,y)
```

```python
# Lasso regression
# Ridge regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso=Lasso()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor=GridSearchCV(lasso,params,X,y,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X,y)
print(lasso_regressor.best_score_)
```
